```python
import pandas as pd
```

```python
data = pd.read_csv('tracks.csv', encoding="utf-8", index_col=False)
display(data.head())
```

```python
duplicates = data['id'].value_counts()[data['id'].value_counts() > 1].index
tmp = data.loc[[n in duplicates for n in data['id']]]
tmp.head()
```

```python
tmp = tmp.sort_values(by=['id', 'idx'])
tmp.head()
```

```python
tmp.to_csv('duplicate.csv', encoding="utf-8", index=False)
```

```python
import numpy as np
import random
```

```python
random.seed(113)

n = len(tmp['id'].unique())

for i, ID in enumerate(tmp['id'].unique()):
    print(f'[{i+1}/{n}]', end='\r')
    remove = tmp[tmp['id'] == ID]['idx'].values
    leave = random.randint(0, len(tmp[tmp['id'] == ID])-1)
    remove = np.delete(remove, leave)
    data = data[[i not in remove for i in data['idx']]]
```

```python
data.to_csv('tracks_no_duplicate.csv', encoding="utf-8", index=False)
```
